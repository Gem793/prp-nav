```python
select_widget=widgets.Select(
    options=['Option A', 'Option B'],
    value='Option A',
    description='Select',
    disabled=False)

def selectOption(opt):
    if opt == 'Option A':
        print('A')
    if opt == 'Option B':
        print('B')

widgets.interact(selectOption, opt=select_widget)
```

```python
class Navigator:
    def __init__(self):
        self.Location = (12.971279989048629, 79.16634675433899)
        self.start = 'ROOM1'  
        self.destination = 'ROOM2' 
        self.geoResources = {
            'ROOM1ROOM2': 'georesources/ROOM1ROOM2.geojson',
            'ROOM1ROOM3': 'georesources/ROOM1ROOM3.geojson',
            'ROOM2ROOM1': 'georesources/ROOM2ROOM1.geojson',
            'ROOM2ROOM3': 'georesources/ROOM2ROOM3.geojson',
            'ROOM3ROOM1': 'georesources/ROOM3ROOM1.geojson',
            'ROOM3ROOM2': 'georesources/ROOM3ROOM2.geojson',
            'ROOM1': 'georesources/ROOM1.geojson',
            'ROOM2': 'georesources/ROOM2.geojson',
            'ROOM3': 'georesources/ROOM3.geojson'
        }
    
    def changeStart(self, newStart):
        self.start = newStart
    
    def changeDestination(self, newDestination):
        self.destination = newDestination
        self.redrawMap()

    def drawRooms(self, Map):
        for room in [self.start, self.destination]:
            geoFile = self.geoResources.get(room)
            if geoFile:
                folium.GeoJson(geoFile, name=room).add_to(Map)
    
    def drawPath(self, Map):
        key = f"{self.start}{self.destination}"
        geoFile = self.geoResources.get(key)
        if not geoFile:
            print(f"No path GeoJSON for {key}")
            return
        
        with open(geoFile) as f:
            geoData = json.load(f)
        
        for feature in geoData['features']:
            path = feature['geometry']['coordinates']
            path = [[c[1], c[0]] for c in path]
            plugins.AntPath(path, color='red', weight=5).add_to(Map)
    
    def redrawMap(self):
        mapprp = folium.Map(location=self.Location, zoom_start=35, width='75%')
        self.drawRooms(mapprp)
        self.drawPath(mapprp)
        display(mapprp)

```

```python
myNavigator = Navigator()
```

```python
def selectStart(room):
    myNavigator.changeStart(room)
```

```python
def selectDestination(room):
    myNavigator.changeDestination(room)
```

```python
start_widget = widgets.Select(
    options=['ROOM1','ROOM2','ROOM3'],
    value='ROOM1',
    description='Start'
)

destination_widget = widgets.Select(
    options=['ROOM1','ROOM2','ROOM3'],
    value='ROOM2',
    description='Destination'
)

# ---------------- Interact ----------------
widgets.interact(selectStart, room=start_widget)
widgets.interact(selectDestination, room=destination_widget)

```
